```python
def PercentReturn (DataFrame,looper):
    Set = DataFrame[DataFrame.ReCata == looper]
    HG_Sound = ((Set['GIF_ReCata'].tolist().count(1))/Set.shape[0])*100
    SG_Sound = ((Set['GIF_ReCata'].tolist().count(2))/Set.shape[0])*100
    
    print("{}: Hard G sound likelyhood is {} percent of the time.".format(looper, HG_Sound))
    print("{}: Soft G sound likelyhood is {} percent of the time.".format(looper, SG_Sound))
    return 
```

```python
for i in range(0,20):
    PercentReturn(YearsCoding,i)
    
```

Based on the data above and by making the slight assumption that the number of years that person has been coding roughly corrosponds to a persons age. We can make the conclusion that age does factor into how likely a somebody is to use either the hard G sounding GIF or the soft G sounding GIF. 

It appears that from the point where somebody coding for 10 years or more they are less likely to use the hard G sound and more likely to pronounce it as "jiff". The chance seems to drop off by roughly 7-10% but does seem to be significant enough to warrant this conclusion.

Conversly, a reletivly younger person seems to have a higher chance of pronouncing the GIF with the hard g sound.
